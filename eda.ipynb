```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/result.csv')
```

```python
df
```

```python
# scores by gender
df[df['Gender'] == 'M'].score.hist(bins=20, alpha=0.2, label='M')
df[df['Gender'] == 'F'].score.hist(bins=20, alpha=0.2, label='F')
df[df['Gender'].isnull()].score.hist(bins=20, alpha=0.2, label='Unknown')
plt.title('Histogram of scores by gender')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
```

```python
# scores by Veteran status
df[df['Veteran status'] == 1].score.hist(bins=20, alpha=0.2, label='Veteran')
df[df['Veteran status'] == 0].score.hist(bins=20, alpha=0.2, label='Non-veteran')
plt.title('Histogram of scores by Veteran status')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
```

```python
# scores by Work authorization
df[df['Work authorization'] == 1].score.hist(bins=20,
                                             alpha=0.2,
                                             label='Authorized')
df[df['Work authorization'] == 0].score.hist(bins=20,
                                             alpha=0.2,
                                             label='Not authorized')
plt.title('Histogram of scores by Work authorization')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
```

```python
# scores by Disability

df[df['Disability'] == 1].score.hist(bins=20, alpha=0.2, label='Disabled')
df[df['Disability'] == 0].score.hist(bins=20, alpha=0.2, label='Not disabled')
df[df['Disability'].isnull()].score.hist(bins=20, alpha=0.2, label='Unknown')
plt.title('Histogram of scores by Disability')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
```

```python
# scores by Ethnicity

df[df['Ethnicity'] == 1].score.hist(bins=20, alpha=0.2, label='Hispanic')
df[df['Ethnicity'] == 0].score.hist(bins=20, alpha=0.2, label='Non-Hispanic')
df[df['Ethnicity'].isnull()].score.hist(bins=20, alpha=0.2, label='Unknown')
plt.title('Histogram of scores by Ethnicity')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
```
